```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

$X \sim Exp(\lambda = 0.25)$

```python
l = 0.25 #Lambda
beta = 1/l #Beta 

X = stats.expon(scale = beta)

x = np.arange(-10,40,0.1)

# FDP - Funcion de densidad de probabilidad (f(x))
pdf = X.pdf(x)
plt.plot(x, pdf)
plt.xlabel('X')
plt.ylabel('Probability')
plt.show()
```
